# Comparação de Previsões e Riscos Reais por Estado

```python
import pandas as pd
import matplotlib.pyplot as plt

# Exemplo de dados reais (df_real)
df_real = pd.DataFrame({
    'Estado': ['SP', 'RJ', 'MG', 'SC', 'RS', 'PR'],
    'Risco Real': [100, 50, 120, 30, 80, 60],  # Quantidade de ocorrências reais de risco por estado
})

# Exemplo de matrizes de confusão (df_previsao)
# Cada estado tem sua respectiva matriz de confusão (verdadeiro positivo, falso positivo, falso negativo, verdadeiro negativo)
df_previsao = pd.DataFrame({
    'Estado': ['SP', 'RJ', 'MG', 'SC', 'RS', 'PR'],
    'TP': [80, 45, 110, 25, 70, 55],  # True Positives
    'FP': [10, 5, 7, 2, 8, 4],        # False Positives
    'FN': [5, 3, 10, 1, 5, 3],        # False Negatives
    'TN': [5, 0, 3, 2, 7, 4]          # True Negatives
})

# Unindo os dados reais com as matrizes de confusão
df_comparacao = pd.merge(df_real, df_previsao, on='Estado')

# Calculando previsões corretas (True Positives + True Negatives)
df_comparacao['Previsões Corretas'] = df_comparacao['TP'] + df_comparacao['TN']

# Calculando previsões incorretas (False Positives + False Negatives)
df_comparacao['Previsões Divergentes'] = df_comparacao['FP'] + df_comparacao['FN']

# Plotando gráfico de comparação
fig, ax = plt.subplots(figsize=(10, 6))

# Plotando barras de previsões corretas e incorretas para cada estado
df_comparacao.set_index('Estado')[['Previsões Corretas', 'Previsões Divergentes']].plot(kind='bar', ax=ax)

for p in ax.patches:
    ax.annotate(f'{p.get_height()}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', fontweight='bold', xytext=(0, 5),
                textcoords='offset points')

# Ajustando o gráfico
plt.title('Comparação entre Previsões e Dados Reais por Estado', fontsize=14)
plt.ylabel('Quantidade de Previsões', fontsize=12)
plt.xlabel('Estado', fontsize=12)
plt.xticks(rotation=0)
plt.tight_layout()

# Exibindo o gráfico
plt.show()

# Exibindo a tabela para conferirmos os dados
print(df_comparacao)
```
